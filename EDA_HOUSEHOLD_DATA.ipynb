# INSIGHTS FROM EDA

*Most of the households in all classes own the house. Very less people live in rental house.

*Some extreme and moderate poverty people pay an equal amount of rent.

*In Non-vulnerable class majority of rooms where above five. While in other classes the majority of rooms where between 2-4.
98% of  all classes had refrigerator.

*Majority of the houses don’t have Television. Even  people having tablet don’t have Television. This shows people are not interested in Television.

*99% of all classes had mobile. Interesting fact is  some extreme poverty people had more mobiles while compared with  other classes.

*Male younger than 12years where high in extreme poverty and non-vulnerable class.

*Total number of males where equally distributed in all classes.

*Female count was less in Non-vulnerable class.

*Children count where high in extreme poverty class and moderate poverty class

*People per room was high in extreme poverty class. Even it exceeded more than 5 in a room. While in other classes it was less compared with extreme poverty people.

*Majority of the extreme poverty class people had  0 years of schooling.

*Non-vulnerable people has best record in schooling.

*As know and expected quality of house is better in vulnerable and non-vulnerable classes while compared with other two classes.
